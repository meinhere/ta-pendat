# Data Understanding

## Pemahaman Data

**Ikhtisar Kumpulan Data:**
Kumpulan data tersebut terdiri dari parameter neurokognitif dan fisiologis pasien, serta tingkat keparahan trauma mereka. Berikut kolom dan penjelasannya:

- **Patinet_ID:** Pengidentifikasi unik untuk setiap pasien.
- **EEG_Signal_Amplitudo:** Amplitudo sinyal EEG.
- **EEG_Delta_band:** Pembacaan EEG pada pita frekuensi delta.
- **EEG_Theta_band:** Pembacaan EEG pada pita frekuensi theta.
- **EEG_Alpha_band:** Pembacaan EEG pada pita frekuensi alfa.
- **EEG_Beta_band:** Pembacaan EEG pada pita frekuensi beta.
- **Fenotipe_Rambut:** Jenis rambut (Rambut_keriting, Rambut_bergelombang, Rambut_Lurus, Rambut_Tanpa).
- **detak_jantung:** Kategori detak jantung (Rendah_Pulse Rendah, Kecepatan_Pulse Sedang, Kecepatan_Pulse Tinggi).
- **konduktansi_kulit:** Kategori konduktansi_kulit (Konduktansi_Rendah, Konduktansi_Normal, Konduktansi_Tinggi).
- **suhu_kulit:** Kategori suhu kulit (Suhu_Rendah, Suhu_Normal, Demam).
- **tingkat_kortisol:** Kategori tingkat kortisol (Di Bawah_AverageCL, Rata-rataCL, Di Atas_AverageCL).
- **Systolic_BP:** Kategori tekanan darah sistolik (Range1_LowSystolic, Range2_LowSystolic, Range3_LowSystolic).
- **Diastolic_BP:** Kategori tekanan darah diastolik (Sangat RendahDiSystolic, LowDiSystolic, NormalDiSystolic).
- **Keparahan_Trauma:** Keparahan trauma (Keparahan_Rendah, Keparahan_Sedang, Keparahan_Tinggi).

# Modelling

## Modelling menggunakan fitur data neurologi

```python
y = data_df["Trauma_Severity"]
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
full_data.iloc[3:4,]
```

```python
full_data.columns[0:5]
```

```python
arr_neuro = full_data[full_data.columns[0:5]].to_numpy()
arr_neuro
```

```python
features = arr_neuro
labels = y

GNB = GaussianNB()

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.4, random_state=7)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []
for train_index, val_index in kf.split(X_train):
    X_train_fold, X_val_fold = X_train[train_index], X_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]
    GNB.fit(X_train_fold, y_train_fold)
    predictions = GNB.predict(X_val_fold)
    accuracy = accuracy_score(y_val_fold, predictions)
    accuracies.append(accuracy)
    print("Akurasi validasi set:", accuracy)

neuro_GNB = GaussianNB()
neuro_GNB.fit(X_train, y_train)
predictions = neuro_GNB.predict(X_test)
test_accuracy = accuracy_score(y_test, predictions)
print("Akurasi test set:", test_accuracy)
print("Rata-rata akurasi semua folds:", sum(accuracies) / len(accuracies))
```

## Evaluation untuk model fitur data neuro

```python
X_test[5]
```

```python
print(classification_report(y_test, predictions, target_names=le.classes_))
```

```python
cm = confusion_matrix(y_test, predictions)
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
